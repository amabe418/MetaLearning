# **Benchmark**

Un **benchmark (como CC18 o CTR23 en OpenML)** te proporciona:

* **Un conjunto estándar de datasets** ya seleccionados y aceptados por la comunidad.
* **Tareas bien definidas** (target, tipo de problema, splits, validación, métricas).
* **Reproducibilidad**: otros pueden repetir exactamente tus experimentos.
* **Comparabilidad científica**: tus resultados se pueden comparar con papers existentes.
* **Calidad y curaduría** de los datos (menos ruido, menos sesgos arbitrarios).

**No te da**:

* Metafeatures calculadas.
* El meta-dataset.
* El meta-learner entrenado.

```python
from config import benchmarks
import openml

for bc in benchmarks:
    print("=" * 80)
    print("Benchmark:", bc)
    bench_cfg = benchmarks[bc]

    assert bench_cfg["type"] == "openml_suite"

    if "suite_id" in bench_cfg:
        suite = openml.study.get_suite(bench_cfg["suite_id"])
    else:
        suite = openml.study.get_suite(bench_cfg["suite_name"])

    print("Suite name:", suite.name)
    print("Num datasets:", len(suite.data))
    print("Num tasks:", len(suite.tasks))

    # Iterar por tasks
    for task_id in suite.tasks[:5]:  
        
        # limita si no quieres miles
        task = openml.tasks.get_task(task_id)
        print("\nTask ID:", task_id)
        print("Dataset ID:", task.dataset_id)

        # Obtener runs de esa task
        runs = openml.runs.list_runs(task=[task_id])

        print("Num runs:", len(runs))

        # Mostrar algunos runs
        for run_id, run in list(runs.items())[:5]:
            print(
                f"  Run {run_id} | "
                f"Flow={run['flow_name']} | "
                f"Uploader={run['uploader']} | "
                f"Date={run['date']}"
            )
```
