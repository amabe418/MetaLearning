```python
import pandas as pd

# Cargar tus dos datasets de meta-features
df1 = pd.read_csv("../notebooks/basic_representations.csv")
df2 = pd.read_csv("meta_features_task2.csv")

# Lista de columnas excluyendo id o nombres si los hay
features1 = set(df1.columns) - {"task_id", "dataset_name"}
features2 = set(df2.columns) - {"task_id", "dataset_name"}


common_features = features1.intersection(features2)
print("Número de meta-features comunes:", len(common_features))
print("Meta-features comunes:")
print(common_features)


only_in_df1 = features1 - features2
only_in_df2 = features2 - features1

print("Solo en dataset1:", only_in_df1)
print("Solo en dataset2:", only_in_df2)
```

```python
# Elegir el task_id que quieres comparar
task_id_to_check = 3

# Filtrar ese task_id en ambos datasets
row1 = df1[df1["task_id"] == task_id_to_check]
row2 = df2[df2["task_id"] == task_id_to_check]

# Si hay más de una fila por task_id, toma la primera
row1 = row1.iloc[0]
row2 = row2.iloc[0]

# Comparar solo las columnas comunes
differences = {}
for feature in common_features:
    val1 = row1[feature]
    val2 = row2[feature]
    if val1 != val2:
        differences[feature] = (val1, val2)

print(f"Meta-features diferentes para task_id={task_id_to_check}:")
for feature, (v1, v2) in differences.items():
    print(f"{feature}: dataset1={v1}, dataset2={v2}")

print(f"Total de meta-features diferentes: {len(differences)}")
print(f"Total de meta-features iguales: {len(common_features) - len(differences)}")
```
