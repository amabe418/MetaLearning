```python
import openml

benchmark = openml.study.get_suite("OpenML100")

dataset_ids = set()
failed_tasks = []

for task_id in benchmark.tasks:
    try:
        task = openml.tasks.get_task(task_id, download_splits=False)
        dataset_ids.add(task.dataset_id)
    except Exception as e:
        failed_tasks.append(task_id)

print("Datasets únicos:", len(dataset_ids))
print(sorted(dataset_ids))
print("Tasks fallidos:", failed_tasks)
```
