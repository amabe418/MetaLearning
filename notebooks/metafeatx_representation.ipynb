```python
import pandas as pd

# Cargar el CSV
df = pd.read_csv("metafeatx_adaboost_representation.csv")

# Contar la cantidad de datasets Ãºnicos
dataset_ids = df['task_id'].values
print("Cantidad de datasets: ", len(dataset_ids))
print(dataset_ids)
```

```python
X= df.drop(columns=['task_id']).values
X
```

```python
from sklearn.metrics import pairwise_distances

D = pairwise_distances(X, metric="euclidean")

D_df = pd.DataFrame(D, index=dataset_ids, columns=dataset_ids)
print(D_df)
```

```python
task=3

distances = D_df.loc[task].sort_values()

print(distances)
```

```python
neighbors = distances.iloc[1:6]  # top-5 vecinos
print(neighbors)
```
